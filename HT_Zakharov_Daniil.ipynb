- Сколько стоит подписка относительно среднего дохода (базовая, стандартная, премиум) в Чехии?

```python
cz_salary = salary[salary['Country'] == 'Czech Republic']['Salary'].iloc[0]
cz_salary
```

```python
cz_basic_perc = df[df['Country'] == 'Czechia']['Cost Per Month - Basic ($)'].iloc[0] * 100 / cz_salary 
cz_standard_perc = df[df['Country'] == 'Czechia']['Cost Per Month - Standard ($)'].iloc[0] * 100 / cz_salary 
cz_premium_perc = df[df['Country'] == 'Czechia']['Cost Per Month - Premium ($)'].iloc[0] * 100 / cz_salary
print(f'Basic cost to average salary in Czechia = {cz_basic_perc} %')
print(f'Standard cost to average salary in Czechia = {cz_standard_perc} %')
print(f'Premium cost to average salary in Czechia = {cz_premium_perc} %')
```

```python
ru_basic_perc = df[df['Country'] == 'Russia']['Cost Per Month - Basic ($)'].iloc[0] * 100 / ru_salary 
ru_standard_perc = df[df['Country'] == 'Russia']['Cost Per Month - Standard ($)'].iloc[0] * 100 / ru_salary 
ru_premium_perc = df[df['Country'] == 'Russia']['Cost Per Month - Premium ($)'].iloc[0] * 100 / ru_salary
print(f'Basic cost to average salary in Russia = {ru_basic_perc} %')
print(f'Standard cost to average salary in Russia = {ru_standard_perc} %')
print(f'Premium cost to average salary in Russia = {ru_premium_perc} %')
```
